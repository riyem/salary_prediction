```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()
```

```python
##now the salary coloumn has Naan values ..we will eliminate them
df=df[df['Salary'].notnull()]
df.head()
```

```python
df['Country'].value_counts()
```

```python
df = df.drop("Employment", axis=1)
df.info()
```

```python
##delete countries with just one data count beacause model can't learn from it anything

def shorten_categories(categories, cutoff):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values [i]>= cutoff:
            categorical_map[categories.index[i]]= categories.index[i]
        else:
            categorical_map[categories.index[i]]='Other'

    return categorical_map

##If the category value is greater than or equal to the cutoff, it assigns the category name to itself in the categorical_map dictionary, effectively leaving it unchanged.

```

```python
country_map=shorten_categories(df['Country'].value_counts(),400)
df['Country']=df['Country'].map(country_map)
df['Country'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(12, 7))

# Group salaries by country
salaries_by_country = df.groupby('Country')['Salary'].mean().sort_values()

# Create horizontal bar chart
ax.barh(salaries_by_country.index, salaries_by_country, color='skyblue')

# Set the title and labels
ax.set_title('Average Salary by Country (US$)', fontsize=16)
ax.set_xlabel('Average Salary (US$)', fontsize=14)
ax.set_ylabel('Country', fontsize=14)

# Show grid
ax.grid(True)

# Show the plot
plt.show()
```

```python
##clean yearsCodePro
df['YearsCodePro'].unique()
```

```python
##we need to convert all the strings and the values: 'Less than 1 year' and 'More than 50 years' to floats
def clean_experience(x):
    if x ==  'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df["EdLevel"].unique()
##we want to combine few categories if it's a pro or doc degree will consider it as postgrad
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
##some of the data is float and other is string we need to transform string values to unique values the model can understand

from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
#le.classes_
```

```python
##we do the same for Country
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
##the label which will be predicted is the salary and all the rest columns are the features 
X = df.drop("Salary", axis=1)
y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y.values)
```

```python
df.describe()
##
#Based on the summary statistics provided, there are NaN values present in the YearsCodePro column. Here's how we can identify this:

#The count row represents the number of non-null values in each column.
#In the YearsCodePro column, the count is less than the total number of entries (36551 < 36680), indicating that there are missing values in that column.
#Therefore, the YearsCodePro column contains NaN values that need to be handled appropriately, such as by imputation or removal.
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in X with the mean
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Instantiate and fit the LinearRegression model
linear_reg = LinearRegression()
linear_reg.fit(X_imputed, y.values)
```

```python
y_pred = linear_reg.predict(X_imputed)
```

```python
#evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred)) #y is the actual label vs the y_pred is the predicted label
```

```python
print("${:,.02f}".format(error))
#our model on avg is off by this number -->very high number
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X_imputed, y.values)
```

```python
y_pred = dec_tree_reg.predict(X_imputed)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)
```

```python
y_pred = random_forest_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
#to find the best model with best parameters we do gridsearch with cross validation
from sklearn.model_selection import GridSearchCV
#set of diff parameters we want to try out
max_depth = [None, 2,4,6,8,10,12]
#create a dictionary
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X_imputed, y.values)
#loop through all the parameters and each time determine the error
```

```python
regressor = gs.best_estimator_

regressor.fit(X_imputed, y.values)
y_pred = regressor.predict(X_imputed)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
